```python
import pandas as pd
import numpy as np

# Sample data for 10 countries 
data = {
    'Country':
    ['Norway', 'Sweden', 'Greece', 'USA', 'India', 'Japan', 'Germany', 'Brazil', 'Canada', 'SouthKorea'],
    'GDP_per_capita': [75000, 55000, 20000, 65000, 2200, 40000, 50000, 15000, 48000, 32000],
    'Happiness': [7.5, 7.2, 5.8, 7.0, 4.0, 6.5, 6.9, 5.0, 7.1, 5.9]
}

df = pd.DataFrame(data)
df
```

```python
# Mean
print("Mean GDP:", df['GDP_per_capita'].mean())
print("Mean Happiness:", df['Happiness'].mean())

# Standard Deviation
print("GDP std:", df['GDP_per_capita'].std())
print("Happiness std:", df['Happiness'].std())
```

```python
# Correlation between GDP and Happiness
corr = df['GDP_per_capita'].corr(df['Happiness'])
print("Correlation (r):", corr)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='GDP_per_capita', y='Happiness', data=df)
plt.title("Happiness vs GDP per Capita")
plt.xlabel("GDP per Capita ($)")
plt.ylabel("Happiness Score")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 1: Prepare the data
X = df[['GDP_per_capita']]  # Feature (predictor)
y = df['Happiness']        # Target (what we want to predict)

# Step 2: Create the model
model = LinearRegression()

# Step 3: Fit the model
model.fit(X, y)

# Step 4: Get slope and intercept
slope = model.coef_[0]
slope_per_1000 = slope * 1000
intercept = model.intercept_
print("Slope per $1000 GDP:", slope_per_1000)
print("Intercept:", intercept)

# Step 5: Calculate RÂ²
r2 = model.score(X, y)
print("RÂ²:", r2)

# Step 6: Optional - plot regression line
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='GDP_per_capita', y='Happiness', data=df)
plt.plot(df['GDP_per_capita'], model.predict(X), color='red')  # regression line
plt.title("Happiness vs GDP per Capita with Regression Line")
plt.xlabel("GDP per Capita ($)")
plt.ylabel("Happiness Score")
plt.show()
```
